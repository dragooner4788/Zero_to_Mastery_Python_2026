```python
# Integers

a = 10

print(type(a)) # 10

isinstance(a, int) # True
```

## Integers Review

As you can see above, we assigned the value `10` to the variable `a`:

`a = 10`

Then use the function `print()` where we use the function to print the value of the variable `a` to the console. 

```python
# Floats

print(float(a)) # 10

b = 10.5

print(b) # 10.5
print(type(b)) # <class 'float'>
print(f"Is variable 'b' a Float? --> {isinstance(b, float)}") # True
```

We can see  a feww things with **Floating Point** Numbers: 

1. We can cast an **Integer** as a **Floating Point** Number. 
2. We can assign a variable a **Floating Point** Number
3. We can check the `type()` of a variable
4. We can use `isinstance(<variable>, <type>)` to see if a specific variable returns True to that specific data type

```python
# Booleans

x = 2
y = 3

print(x > y) # False
print(type(x > y)) # <class 'bool'>
print(isinstance(x > y, bool)) # True  
print(x < y) # True
print(type(x < y)) # <class 'bool'>
print(isinstance(x < y, bool)) # True
```

```python
# Truthy Values

print("TRUTHY VALUES:")
print(bool(1)) # True
print(bool(-1)) # True
print(bool(0.1)) # True
print(bool("Hello")) # True
print(bool([1, 2, 3])) # True
print(bool((1, 2))) # True
print(bool({"key": "value"})) # True
print(bool({1, 2, 3})) # True

print("FALSY VALUES:")
# Falsy Values
print(bool(0)) # False
print(bool(0.0)) # False
print(bool("")) # False
print(bool([])) # False
print(bool(())) # False
print(bool({})) # False
print(bool(None)) # False
print(bool(False)) # False
```
