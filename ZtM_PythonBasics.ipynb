# Math Functions

## Round - round()

There is a function `round()` which is provided one or two parameters. 

If it is just provided the number to round, it will return a rounded number to the nearest Integer. 

If it is provided the number to round, and how many decimal places, it will then round to the nearest decimal place. 

## Absolute Value - abs()

This will provide the absolute value of a given number. 

If we have -5, it will return 5, because -5 is 5 units away from the point of origin which is 0. 

````python
# Math Functions

# Round

print(f"The value of {2.765} when being rounded is: {round(2.765)} using the round() function.")

# Round -  Num Digits

print(f"The value of {2.765} when being rounded is: {round(2.765, 1)} using the round(n, num digits) function where num digitds is 1.")

# Absolute Value
# abs() function

print(f"The absolute valule of -5 is {abs(-5)}")
print(f"The absolute value of {-2.765} is {abs(-2.765)}")
````

# Operator Precedence

I believe Operator Precedence follows PEMDAS - Please Excuse My Dear Aunt Sally

1. Prentheses
2. Exponents
3. Multiplication
4. Dvision
5. Addition
6. Subtraction

````python
# Problem 1 

x = 20
y = 3
z = 4

print(x + y * z) #or 
print(20 + 3 * 4)
````

I wanted to show how Operator Precedence works in Programming and in Math

Problem: 20 + 3 * 4

Process: 

1. (3 * 4) = 13 - We do this first because Multiplication is higher in Operator Precedence than Addition.
2. Problem: 20 + 12
3. We now add 20 and 12 together
4. Answer: 32

**Note**: We can use *Parentheses* to change the order in which we would do the computation. 

````python
# Using Parentheses - which have the highest order to switch the order in which we do the operation. 
# Remember - you always do what is in Parentheses first

x = (20 + 4) * 4 
print(x)

# We could do this for the whole list of Operators...

# Exponents
# 1. 2^2 = 4
# 2. 4 + 4 = 8
#3. Answer: 8

# Wrong Way
# 1. 2 + 4 = 6
# 2. 2^6 = 2 * 2 * 2 * 2 * 2 * 2 = 64

my_exponent = 2 ** 2 + 4

my_exponent_wrong = 2 ** (2 + 4) # although this would be correct because we are using parentheses and thus following the order. But let's pretend we are doing this on paper. 

print(my_exponent)
print(my_exponent_wrong)
````
