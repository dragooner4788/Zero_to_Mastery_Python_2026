# Data Types

## Fundamental Data Types

1. `int`: Integer
2. `float`: Floating Point
3. `bool`: Boolean (True, False)
4. `str`: Strings (Text)
5. `list`: Ordered collection of items, with a position.
6. `tuple`: An immutable collection of items, with a position. 
7. `set`: An ordered collection of item, where it only shows the FIRST instance of an item.
8. `dict`: Key-Value Pairs of data, where *accessing* the key provides the *data*

## Classes | Custom Types

We can create out own Data Types by using Classes.

## Specialized Data Types

These are not built-in to python, but they come from packages which you can use with Python. 

**Modules**: Extensions we can use with the language. 

## None

`None`: This means there is **nothing**. It is like the concept of zero in math. It is the representation of nothing.

```python
# Fundamental Data Types - Numbers

# Integers (int) are whole numbers, positive or negative, without decimals, of unlimited length.

my_integer = 42
print(type(my_integer))  # Output: <class 'int'>
print(f"The variable my_integer has a value of {my_integer} and is of type: {type(my_integer)}")

print("We can do operations with integers such as add, subtract, divid, multiply...")

a = 5
b = 2
x = 7
y = 3

my_sum = a + b # 7

my_diff = a - b # 3

my_product = x * y # 21

my_dividend = 7 / 3 # 2.33

print(f''' 
    The sum of a(5) and b(2) is: {my_sum}.
    The difference of a(5) and b(2) is: {my_diff}.
    The product of x(7) and y(3) is: {my_product}.
    The dividend of x(7) and y(3) is: {my_dividend}.
''')

# Floating-point numbers (float) are real numbers with a decimal point.

my_float = 3.14
print(type(my_float))  # Output: <class 'float'>
print(f"The variable my_float has a value of {my_float} and iks of type: {type(my_float)}")

# Complex numbers (complex) have a real and an imaginary part.
```
