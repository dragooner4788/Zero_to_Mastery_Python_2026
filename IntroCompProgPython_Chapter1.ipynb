A **Computer** does two things:

1. Performs Calculations
2. Remembers the results of those calculations

A **Byte** is 8 **Bits**:

- 1 Byte (8 Bits) represents 1 Character

**Declarative Knowledges**: This is the stating of a **fact**.

- 2 + 2 = 4
- A vehicle transports goods, and people

**Imperative Knowledge**: *How-to* knowledge. It is like a recipe on how to get a result. 

- How to build a Lego set
- How to bake a cake

**Algorith**: A sequence of simple steps with execution flow. 

- The example he gives in the book is how someone found out how to calculate the square root of a number.
- The process is a **guess-and-check** algorithm. 

**Question**: If we are guessing and checking, when do we know we are right?

**Fixed-Program Computer**: This performs a singular task, like a calculator for example. It is used to calculate, but not much else.

**Stored-Program Computer**: This can store a set of instructions. Any set of instructions you come up with in theory should then run.
